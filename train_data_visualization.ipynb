```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Day")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best")
```

```python
plt.xlabel("Day")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
```

```python
company=['GOODL','AMZN','MSFT','F8']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("US TECH STOCKS")
plt.bar(ypos,revenue,label="Revenue")
plt.bar(ypos,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("US TECH STOCKS")
plt.bar(ypos+0.2,revenue,width=0.4,label="Revenue")
plt.bar(ypos-0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)

plt.title("US TECH STOCKS")
plt.barh(ypos+0.2,revenue,label="Revenue")
plt.barh(ypos-0.2,profit,label="Profit")
plt.legend()
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar_men,bins=3)
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar_men,bins=[80,100,125,200],rwidth=0.95,color='g',histtype='step')
```

```python
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","car","Other Faculties"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True,explode=[0,0.5,0,0,0],startangle=45)
plt.show()
```
